```python
import matplotlib.pyplot as mp
%matplotlib inline
```

```python
def analyze(dataset):
    # Create canvas for the plot
    fig = mp.figure(figsize=(10,3))

    # Earmark spots on the canvas for the plot
    axes1 = fig.add_subplot(1,3,1) ## create subplot
    axes2 = fig.add_subplot(1,3,2)
    axes3 = fig.add_subplot(1,3,3)

    axes1.plot(dataset.mean(axis=0)) ## plot the average
    axes1.set_ylabel('average') ## set the ylabel

    axes2.plot(dataset.max(axis=0))
    axes2.set_ylabel('maximum')

    axes3.plot(dataset.min(axis=0))
    axes3.set_ylabel('minimum')
    
    #Display plot
    fig.tight_layout()
    mp.show()
```

```python
analyze(data)
```

```python
analyze(data3)
```

```python
# create analyze1 that takes two parameters: dataset and graph, 
#where graph is a string with the values ['min', 'max','average', 'sd']
# output is a single graph output 
# a - is the parameter for the row/column (aka by patient or day in the inflammation datasets)
def analyze_one(dataset, graph, a):
    import matplotlib.pyplot as mp
    %matplotlib inline

    if graph=='max':
        mp.plot(dataset.max(axis=a))
    elif graph=='min':
        mp.plot(dataset.min(axis=a))
    elif graph=='average':
        mp.plot(dataset.mean(axis=a))
    elif graph=='sd':
        mp.plot(dataset.std(axis=a))        

    mp.show()
```

```python
analyze_one(data, 'min', 0)
```

```python
def analyze_o(filename, graph, a=0):
    import numpy
    dataset=numpy.loadtxt(filename,delimiter=',')
    import matplotlib.pyplot as mp
    %matplotlib inline

    if graph=='max':
        mp.plot(dataset.max(axis=a))
    elif graph=='min':
        mp.plot(dataset.min(axis=a))
    elif graph=='average':
        mp.plot(dataset.mean(axis=a))
    elif graph=='sd':
        mp.plot(dataset.std(axis=a))        

    mp.show()
```

```python
analyze_o('inflammation-01.csv','min')
```

```python
fnames =["inflammation-01.csv", "inflammation-03.csv", 'inflammation-08.csv']
for x in fnames:
    analyze_o(x, 'max', 1)
```
